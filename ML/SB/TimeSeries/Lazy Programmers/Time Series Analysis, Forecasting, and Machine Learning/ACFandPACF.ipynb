```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
x0=np.random.randn(1000)
```

```python
plt.plot(x0)
```

```python
fig , ax=plt.subplots(figsize=(10,8))
plot_pacf(x0,ax=ax);
```

```python
x1=[0]
for i in range (1000):
    #0.5 * previous value + Standard Deviation * Gausion noise
    #series depends on 1 past value i.e x1[-1] ->AR(1) process
    x=0.5 * x1[-1] + 0.1 * np.random.randn()
    x1.append(x)

x1=np.array(x1)
```

```python
plt.plot(x1)
```

```python
fig , ax=plt.subplots(figsize=(10,8))
plot_pacf(x1,ax=ax);
```

```python
x2=[0,0]
for i in range (1000):
    #0.5 * previous value + Standard Deviation * Gausion noise
    #series depends on 2 past value i.e x1[-1] ->AR(1) process
    #coefficuent for Lag-1 value (x2[-1]) is 0.5
    #coefficuent for Lag-2 value (x2[-2]) is 0.3
    x = 0.5 * x2[-1] - 0.3 * x2[-2]+ 0.1 * np.random.randn()
    x2.append(x)

x2=np.array(x2)
plt.plot(x2)

fig , ax=plt.subplots(figsize=(10,8))
plot_pacf(x2,ax=ax);
```

```python
x5=[0,0,0,0,0]
for i in range (1000):
    #0.5 * previous value + Standard Deviation * Gausion noise
    #series depends on 2 past value i.e x1[-1] ->AR(1) process
    #coefficuent for Lag-1 value (x2[-1]) is 0.5
    #coefficuent for Lag-2 value (x2[-2]) is 0.3
    x = 0.5 * x5[-1] - 0.3 * x5[-2] -0.6 * x5[-5]+ 0.1 * np.random.randn()
    x5.append(x)

x5=np.array(x5)
plt.plot(x5)

fig , ax=plt.subplots(figsize=(10,8))
plot_pacf(x5,ax=ax);
```

```python
#Generate MA(1) process

# Sample mean 0 with standard Deviation 1 from Normal Distribution
errors = 0.1 * np.random.randn(1000)
ma1 = []
'''
Then we'll enter a loop that goes for 1000 iterations, the same size as our errors array.
'''
for i in range(1000):

    '''
    Inside the loop, we'll check whether the Loop iteration index is greater than or equal to one or not.
    If it's not, then at zero, which is a special case for AI, greater than or equal to one, we just
    have the normal MS1 equation.
    '''
    if i >= 1:

        '''
            As you recall, the MA(1) process is generated by adding the current error,
            plus some coefficient times the previous errors.
        '''
        '''
            Again, I've chosen the coefficient 0.5 arbitrarily, so the current value in the Time series
            is 0.5 times the previous error, plus the current error at Index I.
        '''
        x = 0.5 * errors[i-1] + errors[i]
    else:

        '''
        Note that when is equal to zero, there is no past value.
        Therefore, it's not possible to add any previous error.
        Therefore, for the first iteration, X is just equal to the first error.
        '''
        x = errors[i]
    ma1.append(x)
ma1=np.array(ma1)
```

```python
plt.plot(ma1)
```

```python
fig , ax=plt.subplots(figsize=(10,8))
plot_acf(ma1,ax=ax);
```

```python
errors = 0.1 * np.random.randn(1000)
ma2 = []
for i in range(1000):
    x = 0.5 * errors[i-1] - 0.3 * errors[i-2] + errors[i]
    ma2.append(x)
ma2 = np.array(ma2)

plt.plot(ma2);

fig, ax = plt.subplots(figsize=(10, 5))
plot_acf(ma2, ax=ax);
```
