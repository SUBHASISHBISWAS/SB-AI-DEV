```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('datasetExample.csv')
```

```python
data.head(11)
```

```python
data.info()
```

```python
#Handle duplicate rows 
data.drop_duplicates(inplace=True)
data
```

```python
#Reset Index
# Strategy 1 - Create a index column and set the same in the dataframe
indexCol = np.array(list(range(0,data.count()[0])))
data.set_index(indexCol , inplace=True)
data
```

```python
#Duplicate Columns
data.drop(['Age_Group.1'] , axis = 1 , inplace=True)
data
```

```python
# Identify the columns
#CustomerID ---> Continous Numeric
#Age_Group ----> Categorical 
#Rating(1-5)---> Ordinal
#Hotel --------> Categorical
#FoodPreference-> Categorical
#Bill ----------> Continous Numeric
#NoOfPax -------> Continous Numeric
#EstimatedSalary-> Continous Numeric
```

```python
#Handle Numerical Columns
#CustomerID ---> Continous Numeric ---> No need to Handle
#Bill ----------> Continous Numeric
#NoOfPax -------> Continous Numeric
#EstimatedSalary-> Continous Numeric
```

```python
#Lets Check Customer ID
data[data['CustomerID'] < 0]
#If there exists any negative value, replace the same with NaN
```

```python
#Bill ----------> Continous Numeric
#If there exists any negative value, replace the same with NaN
#Use loc
data.Bill.loc[(data.Bill < 0)] = np.nan
```

```python
data
```

```python
#NoOfPax -------> Continous Numeric
data.NoOfPax.loc[(data.NoOfPax < 0)] = np.nan
data
```

```python
#EstimatedSalary-> Continous Numeric
data.EstimatedSalary.loc[(data.EstimatedSalary < 0)] = np.nan
data
```

```python
#Handling Categorical Columns
#Hotel --------> Categorical
#FoodPreference-> Categorical
#Age_Group ----> Categorical 
```

```python
#Hotel --------> Categorical
data.Hotel.unique()
```

```python
#Hotel --------> Categorical
#Lets decide the standard
# Ibis
# LemonTree
# RedFox
# Misspelling Handling ---> Manual Intervention or NLP
# Two Step Action:
# 1. If you are aware(NLP aware), correct the spelling
# 2. If not aware, replace by NaN
```

```python
#Use mask
data['Hotel'].mask(data.Hotel == 'Ibys' , 'Ibis' , inplace=True)
data
```

```python
#FoodPreference-> Categorical
#Veg or Non-Veg
#Our standard will be
#1. Veg
#2. Non-Veg

data.FoodPreference.unique()
```

```python
#Replace
data.FoodPreference.replace(['veg','Vegetarian'] , 'Veg' , inplace=True)
```

```python
data.FoodPreference.replace(['non-Veg'] , 'Non-Veg' , inplace=True)
data
```

```python
#Age_Group ----> Categorical 
data.Age_Group.unique()
```

```python
#Rating(1-5)---> Ordinal
data['Rating(1-5)'].unique()
```

```python
#My rule as a DS is to make NaN
data['Rating(1-5)'].loc[(data['Rating(1-5)'] < 0) | (data['Rating(1-5)'] > 5) ] = np.nan
data
```
