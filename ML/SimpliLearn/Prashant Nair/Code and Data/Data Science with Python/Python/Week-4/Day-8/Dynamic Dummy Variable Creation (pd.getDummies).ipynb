```python
#Create a function that can handle DummyVariables
def getDummyVariables(uniqueList, categorical_value):
    listToCheck = sorted(uniqueList)
    resultVariable = np.zeros(len(listToCheck))
    counter = 0
    
    for i in range(0, len(listToCheck)):
        if listToCheck[i] == categorical_value:
            resultVariable[counter] = 1
        else:
            counter += 1
    
    return resultVariable
    
```
