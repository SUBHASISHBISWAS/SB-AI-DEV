### Use-case:  You have been a dataset containing features that can identify whether the given Iris flower belongs to Setosa, Versicolor and Virginica species. Your job is to perform EDA over the data to make a layman understand the data and perform Inferential statistics to predict whether the given flower belongs to which species.

```python
#Minimum packages you will need to perform EDA

import numpy as np # Linear Algebra and working with array
import pandas as pd #Data Loading, Exploration and Analysis
import matplotlib.pyplot as plt #Data Visualization
import seaborn as sns # Data Visualization --- Enable graph to be displayed in Jupyter notebook
%matplotlib inline  
```

```python
#Lets load Iris dataset
iris = pd.read_csv('Iris.csv')
```

```python
iris.head()
```

```python
#To check if there exists any missing data
iris.info()
```

```python
#Since there exists no missing data, therefore we can start with statistical analysis
```

```python
#Step1: Identify the type of DATA for each column given in the dataset
# SepalLengthCm    150 non-null float64 ------> Numerical ----> Quantitative ---> histogram
# SepalWidthCm     150 non-null float64 ------> Numerical ----> Quantitative ---> histogram
# PetalLengthCm    150 non-null float64 ------> Numerical ----> Quantitative ---> histogram
# PetalWidthCm     150 non-null float64 ------> Numerical ----> Quantitative ---> histogram
# Species          150 non-null object -------> Categorical----> Qualitative ----> Pie, Bar
```

```python
#Qualitative Descriptive Analysis
#Summarizing

iris.Species.value_counts()

#The given data has 3 species namely Iris-virginica, Iris-versicolor and Iris-setosa. The sample we received is a Balanced Sample
# Balanced Dataset !!!
```

```python
# Visualizing Species
iris.Species.value_counts().plot(kind="bar")
```

```python
iris.Species.value_counts().plot(kind="pie")
```

```python
#Quantitative Data
# SepalLengthCm    150 non-null float64 ------> Numerical ----> Quantitative ---> histogram
# SepalWidthCm     150 non-null float64 ------> Numerical ----> Quantitative ---> histogram
# PetalLengthCm    150 non-null float64 ------> Numerical ----> Quantitative ---> histogram
# PetalWidthCm     150 non-null float64 ------> Numerical ----> Quantitative ---> histogram
```

```python
#Get the summary statistics of the Quantitative data
#Measures of Central Tendency
#Measures of Dispersion 

iris.describe() # Quantitative data ---> Omitted Species (non - numerical data)
```

```python
#Viz each column
iris.SepalLengthCm.plot(kind="hist")
```

```python
#Viz each column
iris.SepalWidthCm.plot(kind="hist")
```

```python
#Viz each column
iris.PetalLengthCm.plot(kind="hist")
```

```python
#Viz each column
iris.PetalWidthCm.plot(kind="hist")
```

```python
# Features ---> SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWidthCm
# Label -----> Species
# Classification because your label is categorical data !!!!
```

```python
#Pair Plot with respect to features and labels
#Seaborn
# Regression
#X axis --- feature column
#Y axis --- label

sns.pairplot(iris, hue="Species", size=3)
```

```python
#Relationship of SepalLengthCm and PetalWidthCm for Species
#scatter (Xaxis, Y axis)

fig = iris[iris.Species == 'Iris-setosa'].plot(kind='scatter' ,
         x = 'SepalLengthCm',
         y = 'PetalWidthCm',
         color = 'blue',
        label = 'Setosa'
                                              
         )
iris[iris.Species == 'Iris-versicolor'].plot(kind='scatter' ,
         x = 'SepalLengthCm',
         y = 'PetalWidthCm',
         color = 'orange' ,label = 'Versicolor',
        ax = fig 
         )
iris[iris.Species == 'Iris-virginica'].plot(kind='scatter' ,
         x = 'SepalLengthCm',
         y = 'PetalWidthCm',
         color = 'green',label = 'Virginica',
         ax = fig                                    
         )


fig = plt.gcf() #Subplot

```
