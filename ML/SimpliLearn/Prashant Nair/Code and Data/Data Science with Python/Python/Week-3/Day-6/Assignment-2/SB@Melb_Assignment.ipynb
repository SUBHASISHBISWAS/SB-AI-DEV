```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("melb_data.csv")
```

```python
df
```

```python
df.info()
```

```python
'''
Feture: Car Parking availability, Landsize, BuildingArea, YearBuilt
Labe : Price
'''
```

```python
fetures=df.iloc[:,[12,13,14,15]]
label=df.iloc[:,[4]]
```

```python
fetures.info()
```

```python
label.info()
```

```python
#NO Arrays
npFeatures=fetures.values
npLabel=label.values
```

```python
from sklearn.impute import SimpleImputer

#Car
carImputer =SimpleImputer(missing_values=np.nan,strategy="median")
npFeatures[:,[0]]=carImputer.fit_transform(npFeatures[:,[0]])

#Landsize
landsizeImputer =SimpleImputer(missing_values=np.nan,strategy="median")
npFeatures[:,[1]]=landsizeImputer.fit_transform(npFeatures[:,[1]])


#BuildingArea
buildingAreaImputer =SimpleImputer(missing_values=np.nan,strategy="median")
npFeatures[:,[2]]=buildingAreaImputer.fit_transform(npFeatures[:,[2]])


#YearBuilt
yearBuiltImputer =SimpleImputer(missing_values=np.nan,strategy="median")
npFeatures[:,[3]]=yearBuiltImputer.fit_transform(npFeatures[:,[3]])
```

```python
fetureDF=pd.DataFrame(npFeatures)
fetureDF.info()
```

```python
labelDF=pd.DataFrame(npLabel)
labelDF.info()
```

```python
finalDF=pd.concat([fetureDF,labelDF],axis=1)
finalDF.columns = ['Car','Landsize','BuildingArea','YearBuilt','Price']
finalDF.head()
```

```python
finalDF.drop_duplicates()
finalDF.info()
```
