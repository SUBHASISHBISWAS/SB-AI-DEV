```python
img=cv2.imread('./DATA/00-puppy.jpg')
img.shape
```

```python
#Show image
'''
It's showing the image in kind of a weird bluish tint.

And the reason for that is because open CV and matplotlib.
Expect different orders of the red, green, blue channels.

So the way a map plot lib thinks of image for a red, green, blue

However, open CV What it does is it reads them in, in a slightly different order.
It reads them in as blue, green, red.

So what happened is when we actually read in that file with open CV, we're reading it in as if it was
a blue, green, red channelled image and mapplotlib is displaying it as if it's a red, green, blue image.

So that's why when we actually display the image after reading it in with  CV to read, we get this kind
of weird discoloration.
'''
plt.imshow(img)
```

```python
'''
So what we need to do is actually transform this to go from blue, green, red to red, green, blue,
the way a map plot lib expects it.

So CV two, it actually has functionality for that.

So all we're saying is take this original image and you read it in as a blue, green, red channel,image.
Convert it, please, to red, green, blue.
'''
fix_image=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_image)
```

```python
'''
imagine you have a colored image, but you immediately
want to read it in as a grayscale image.

Well, OpenCV can actually do that for you as well.
'''
img_grey=cv2.imread("./DATA/00-puppy.jpg",cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img_grey,cmap='gray')
```

```python
'''
So, for example, magma is one that would take dark values close to zero and treat them as kind of
a dark, blackish purple and then take light values as kind of an orange.
'''
plt.imshow(img_grey,cmap='magma')
```

```python
'''
I want to talk about resizing images and then also kind of flipping images around and saving image files.

So that's a little confusing at first that it takes in the X and Y kind of the opposite way that the
numpy thinks about it

So if I want to resize the image, I can actually do it for simple resize command.
'''
new_img=cv2.resize(fix_image,(1000,400))
```

```python
plt.imshow(new_img)
```

```python
'''
But this is how Resize works if you want to do it by a particular ratio.
So it's the image you're going to resize.
'''
w_ratio=0.5
h_ratio=0.5 # 50% of orizinal height

new_img=cv2.resize(fix_image,(0,0),fix_image,w_ratio,h_ratio)
```

```python
plt.imshow(new_img)
```

```python
'''
show you how to flip images along the
horizontal or vertical axis.
'''
new_img=cv2.flip(fix_image,0)
plt.imshow(new_img)
```

```python
'''
If you want to flip it along the vertical middle axes, then you just need to say one here.
'''
new_img=cv2.flip(fix_image,1)
plt.imshow(new_img)
```

```python
new_img=cv2.flip(fix_image,-1)
plt.imshow(new_img)
```

```python
'''
Let's also quickly show you how you can save an image file.
'''
cv2.imwrite("./DATA/new_puppyimage.jpeg",fix_image)
```
