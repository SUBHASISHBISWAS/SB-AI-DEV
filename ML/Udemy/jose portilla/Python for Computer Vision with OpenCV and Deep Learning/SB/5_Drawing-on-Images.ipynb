```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
'''
Next we're going to say is create a blank image and the blank image is going to be NP zeros and we're
going to give it a simple shape of.512 by 512.
And then we're going to give it three color channels.
'''
blank_img = np.zeros(shape=(512,512,3), dtype=np.int16)
```

```python
plt.imshow(blank_img)
```

```python
'''
And we're going to start off by drawing a rectangle on this image.

And then what's going to happen is it's going to accept two points

So pt1, that's going to be the vertex of the rectangle.
So the first vertex, essentially the top left corner and pt2 is going to be the bottom right corner.

And then we can assign a color and the color is just going to be an RGB code.

It's actually going to act directly onto the image.
So this is kind of a permanent marker that you're going to draw on the image, so keep that in mind.
'''

cv2.rectangle(blank_img,pt1=(384,10),pt2=(500,150),color=(0,255,0),thickness=10)
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,pt1=(200,200),pt2=(300,300),color=(0,255,0),thickness=10)
plt.imshow(blank_img)
```

```python
'''
So for circles, it's a very similar process we say to circle and then you pass in the image.
'''
cv2.circle(img=blank_img,center=(100,100),radius=50,color=(255,0,0),thickness=8)
plt.imshow(blank_img)
```

```python
'''
passing in a negative number into the thickness, we'll
say negative one thickness.
'''
cv2.circle(img=blank_img,center=(400,400),radius=50,color=(255,0,0),thickness=-1)
plt.imshow(blank_img)
```

```python
'''
Then I want to show you how to draw lines and text,
'''
cv2.line(blank_img,pt1=(0,0),pt2=(512,512),color=(102,255,255),thickness=5)
plt.imshow(blank_img)
```

```python
'''
So sometimes you're going to have to have a custom polygon to wrap around the particular shape.

So here we have four pairs.
Each of these is going to contain an X and Y coordinate.

So if you take a look at the shape here, it's currently in a four by two and we need to reshape this
to be in another set of dimensions.


'''

blank_img = np.zeros(shape=(512,512,3), dtype=np.int16)
vertices=np.array([[100,300],[200,200],[400,300],[200,400]],dtype=np.int32)
vertices.shape
```

```python
'''
So what you do is once you create this array of points, we're going to basically cast it into three
dimensions for those color channels.

what that has essentially done
is added a third dimension, which is just one.
'''
pts=vertices.reshape((-1,1,2))
pts.shape
```

```python
cv2.polylines(blank_img,[pts],isClosed=True,color=(255,0,0),thickness=5)
plt.imshow(blank_img)
```
