```python
#You can also instantiate a tensor containing byte values by calling torch. ByteTensor. A byte is an unsigned integer holding values between 0 and 255. 
# Any value that is not in this range is expressed relative to 255, for example, 261 is 5 and -5 is 251. 
tensor_byte=torch.ByteTensor([0,261,1,-5])
tensor_byte
```

```python
# If you want to initialize a tensor containing all ones, you can use the torch. one's function. This is a 1D tensor with 10 values. 
tensor_ones=torch.ones(10)
tensor_ones
```

```python
#Similarly, you can pass in dimensionality and shape information to torch. zeros and get a tensor of that shape initialized with all zeros.
tensor_zeros=torch.zeros(10)
tensor_zeros
```

```python
#Torch. eye will give you a square matrix and the main diagonal of that square matrix will be all ones as you see on screen. 

tensor_eye=torch.eye(3)
tensor_eye
```

```python
# In order to find the index positions of all the elements within 
# a tensor that are nonzero, you can use the torch.

non_zero=torch.nonzero(tensor_eye)
non_zero
```

```python
# If you want to create a tensor of ones with the same shape as another tensor, you can call the torch. ones_like function. 
# Here we have a tensor of ones with the same shape as our tensor_eye.

tensor_ones_shape_eye=torch.ones_like(tensor_eye)
tensor_ones_shape_eye
```

```python
# Tensors in PyTorch support two kinds of operations, in-place operations and out-of-place operations. 
# Out-of-place operations generate a new tensor. In-place operations modify the same tensor. 
# In this example here, we've initialized a 3 by 3 tensor and we'll call the fill_ operation on the tensor. 
# All in-place operations on PyTorch tensors have the _ suffix. That's how you identify them. We want to fill this tensor with threes. 
# PyTorch tensors support the in-place fill operation. 

initial_tensor=torch.rand(3,3)
initial_tensor.fill_(3)
```

```python
#Let's perform a mathematical operation on the tensor that we initialized. We call initial_tensor. add and pass in 4. 
# This gives us a brand-new tensor, this is an out-of-place operation where all of the values are 7. 
# Every element of the initial_tensor has the value 4 added and we get this resulting tensor that is stored in the new tensor variable. 
# The initial_tensor remains unchanged.

new_tensor=initial_tensor.add(4)
new_tensor
```

```python
#Instead of performing the out-of-place add operation, let's say we perform an in-place operation, add_5 to the initial_tensor, this changes the values in the initial tensor so all values are now 8. 

initial_tensor.add_(5)
```

```python
#Once you have a PyTorch tensor, you can create a view of a tensor with a different shape. 
# We have the initial tensor. We want to create a new view with just 1 dimension and the shape of 6. 
# Our resize tensor here holds the same values as the initial tensor, but in 1D form. 
# You can think of this as a vector view of our original 2D tensor. When you clear the view of a tensor, 
# it essentially points to the same memory as the original tensor, so if you update an element in the original tensor, 
# you'll find that the corresponding element is updated in the view as well.
Image('/Users/subhasish/Documents/APPLE/SUBHASISH/Development/GIT/Interstellar/SB-AI-DEV/DL/Pluralsight/Janani Ravi/Building Deep Learning Models Using PyTorch/SB/Image/2023-03-08_18-04-37.jpg')
```

```python
#When you create a resized view of a tensor, you can also have PyTorch infer the shape of a particular dimension by specifying -1. 
# So we've specified a second dimension to be 2, and PyTorch based on the number of elements in that tensor has inferred that 
# the first dimension should be 3. Reshaping the elements of a tensor are successful only if 
# the number of elements in that tensor is compatible with your reshape operation. 

resized_tensor=initial_tensor.view(-1,3)
resized_tensor.shape
```

```python
#We can call torch. sort to sort the values in this tensor. You can see that by default each row of your 
# original tensor has been sorted by its values. The values in each row here are in ascending order. 
# The torch. sort operation returns the sorted tensor as the less sorted indices.

sorted_tensor,sorted_indices=torch.sort(initial_tensor)
sorted_tensor
```

```python
sorted_indices
```

```python
# If you want to change the dimensions along which we sort, we can specify the dimension as an input argument. 
# Dim is equal to 0 will sort the values in our tensor. Column wise,
sorted_tensor,sorted_indices=torch.sort(initial_tensor,dim=0)
sorted_tensor
```
